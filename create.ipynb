```python
# Create a list of random data
random_data = []

for _ in range(10000):
    name = random.choice(data['Name'])
    location = random.choice(indian_cities[0])
    policy = random.choice(policies)
    add_on_list = random.choice(add_ons)
    category = random.choice(data['Category'])
    industry = random.choice(data['Industry'])
    income = random.randint(min_income, max_income)
    random_data.append([name, location ,policy, add_on_list, category, industry, income])

# Create a Pandas DataFrame
df = pd.DataFrame(random_data, columns=['Name', 'Location', 'Policy', 'Add-ons', 'Category', 'Industry', 'Sum Insured'])

# Display the resulting DataFrame
df.head()
```
